```python
'''Q1 ) Six points with the following attributes are given, calculate and find out clustering representations and dendrogram using Single,
complete, and average link proximity function in hierarchical clustering technique.'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.cluster.hierarchy as shc
from scipy.spatial.distance import squareform, pdist

x = np.array([0.4005,0.2148,0.3457,0.2652,0.0789,0.4548])
y = np.array([0.5306,0.3854,0.3156,0.1875,0.4139,0.3022])

p = ['P1','P2','P3','P4','P5','P6']
d = pd.DataFrame({'Point':p, 'x cordinate':x, 'y cordinate':y})
data = d.set_index('Point')
data
```

```python
rst = pd.DataFrame(squareform(np.round(pdist(data[['x cordinate', 'y cordinate']]),4), 'euclidean'), columns=data.index.values, index=data.index.values)
rst
```

```python
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram using single link proximity function")  
dendrogram_S = shc.dendrogram(shc.linkage(data[['x cordinate', 'y cordinate']], method='single'), labels=data.index)
```

```python
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram using complete link proximity function")  
Dendrogram_C = shc.dendrogram(shc.linkage(data[['x cordinate', 'y cordinate']], method='complete'), labels=data.index)
```

```python
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram using Average link proximity function")  
dendrogram_A = shc.dendrogram(shc.linkage(data[['x cordinate', 'y cordinate']], method='average'), labels=data.index)
```

```python
#importing libraries from sklearn for StandardScaler, AgglomerativeClustering and normalize functions.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing,metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
# importing waring to set filterwarnings to ignore warnings.
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/CC GENERAL.csv')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
dfd = df.drop(['CUST_ID'], axis=1)
dfd.head()
```

```python
dfd.isnull().any()
```

```python
dfd.fillna(df.mean(), inplace=True)
dfd.isnull().any()
```

```python
dfd.corr().style.background_gradient(cmap="Blues")
```

```python
c = dfd.iloc[:,0:-1]
d = dfd.iloc[:,-1]


scaler = preprocessing.StandardScaler()
scaler.fit(c)
C_scaled_array = scaler.transform(c)
C_scaled_df = pd.DataFrame(C_scaled_array, columns = c.columns)
```

```python
C_normalized = preprocessing.normalize(C_scaled_df)
C_normalized = pd.DataFrame(C_normalized)
```

```python
pca_2 = PCA(n_components=2)
pC = pca_2.fit_transform(C_normalized)

pDf = pd.DataFrame(data = pC, columns = ['P1', 'P2'])

fDf = pd.concat([pDf, df[['TENURE']]], axis = 1)
fDf.head()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(fDf['P1'],fDf['P2'],c=fDf['TENURE'],cmap='prism', s =5)
plt.xlabel('pc1')
plt.ylabel('pc2')
```

```python
plt.figure(figsize=(7,7))
plt.scatter(fDf['P1'],fDf['P2'],c=fDf['TENURE'],cmap='prism', s =5)
plt.xlabel('pc1')
plt.ylabel('pc2')
```

```python
agcl = AgglomerativeClustering(n_clusters = 2)
 
# Visualize the cluster
plt.figure(figsize =(6, 6))
plt.scatter(pDf['P1'], pDf['P2'],
           c = agcl.fit_predict(pDf), cmap ='rainbow')
plt.show()
```

```python
agcl3 = AgglomerativeClustering(n_clusters = 3)
 
# Visualizing the clustering
plt.figure(figsize =(6, 6))
plt.scatter(pDf['P1'], pDf['P2'],
           c = agcl3.fit_predict(pDf), cmap ='rainbow')
plt.show()
```

```python
agcl4 = AgglomerativeClustering(n_clusters = 4)
 
# Visualize the cluster
plt.figure(figsize =(6, 6))
plt.scatter(pDf['P1'], pDf['P2'],
           c = agcl4.fit_predict(pDf), cmap ='rainbow')
plt.show()
```

```python
agcl5 = AgglomerativeClustering(n_clusters = 5)
 
# Visualize the cluster
plt.figure(figsize =(6, 6))
plt.scatter(pDf['P1'], pDf['P2'],
           c = agcl5.fit_predict(pDf), cmap ='rainbow')
plt.show()
```

```python
q = [2, 3, 4, 5]
 
# Append the scores of the different models to one score list
silhouette_scores = []
silhouette_scores.append(
        silhouette_score(pDf, agcl.fit_predict(pDf)))
silhouette_scores.append(
        silhouette_score(pDf, agcl3.fit_predict(pDf)))
silhouette_scores.append(
        silhouette_score(pDf, agcl4.fit_predict(pDf)))
silhouette_scores.append(
        silhouette_score(pDf, agcl5.fit_predict(pDf)))

 
# Plot a bar graph to differentiate the results
plt.bar(q, silhouette_scores)
plt.xlabel('Number of clusters', fontsize = 20)
plt.ylabel('S(i)', fontsize = 20)
plt.show()
```
